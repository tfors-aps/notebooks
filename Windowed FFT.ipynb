### Heinzel
Heinzel, G., Rüdiger, A., & Schilling, R. (2002). Spectrum and spectral density estimation by the Discrete Fourier
￼￼￼￼￼￼￼￼￼￼￼￼￼￼￼￼￼transform (DFT), including a comprehensive list of window functions and some new flat-top windows.

http://pubman.mpdl.mpg.de/pubman/faces/viewItemOverviewPage.jsp?itemId=escidoc:152164

```python
fs = 131072.   # same sample frequency as HP35670A
N = 4096       # same as HP3458A
fres = fs/N    # FFT frequency resolution
fnyq = fs/2.   # Nyquist
```

### Synthesis

First, synthesize a signal with two sinusoids both periodic within time window so no leakage should occur.

```python
f1 = 25*fres             # 800 Hz
v1 = 1.0 * np.sqrt(2)    # 1.0 Vrms

f2 = 625*fres            # 20 kHz
v2 = 20e-6 * np.sqrt(2)  # 20.0 uVrms

dt = 1./fs
t = np.array(range(0, N)) * dt
signal = v1 * np.sin(2*np.pi*f1*t) + v2 * np.sin(2*np.pi*f2*t)

plt.plot(t, signal)
plt.xlabel('Time (s)')
plt.ylabel('Signal (V)');
```

### Compute amplitude spectrum according to Heinzel paper with and without windows.

```python
heinzel = {}

# no window
Y = np.fft.fft(signal) / N
f = np.fft.fftfreq(N, dt)
mask = (f > 0) & (f < fnyq)             # frequency mask
fft = 2. * np.abs(Y[mask]) / np.sqrt(2) # RMS amplitude spectrum
pwr = fft**2                            # power spectrum
psd = pwr / fres                        # power spectral density
f = f[mask]
heinzel['none'] = {'f': f, 'fft': fft, 'pwr': pwr, 'psd': psd}

# flattop window
win = scipy.signal.windows.flattop(N)
s1 = win.sum()
s2 = (win**2).sum()
enbw = fs * (s2 / (s1**2))
Y = np.fft.fft(signal*win) / s1
fft = 2. * np.abs(Y[mask]) / np.sqrt(2)
pwr = fft**2
psd = pwr / enbw
heinzel['flattop'] = {'f': f, 'fft': fft, 'pwr': pwr, 'psd': psd}

# hanning window
win = scipy.signal.windows.hanning(N)
s1 = win.sum()
s2 = (win**2).sum()
enbw = fs * (s2 / (s1**2))
Y = np.fft.fft(signal*win) / s1
fft = 2. * np.abs(Y[mask]) / np.sqrt(2)
pwr = fft**2
psd = pwr / enbw
heinzel['hanning'] = {'f': f, 'fft': fft, 'pwr': pwr, 'psd': psd}
```

### Compute using sddsfft with and without windows

```python
sdds = {}

# write discrete-time data to sdds file
with open('time.sdds', 'w') as f:
    f.write('SDDS1\n')
    f.write('&column name="time", units="s", type="double" &end\n')
    f.write('&column name="voltage", units="V", type="double" &end\n')
    f.write('&data mode="binary" &end\n')
    f.write(pack('<l', N)) # num rows
    for r in xrange(N):
        f.write(pack('<d', t[r]))      # time
        f.write(pack('<d', signal[r])) # voltage

# run sddsfft commands
!sddsfft time.sdds none.sdds -col=time,voltage -psdOutput
!sddsfft time.sdds flattop.sdds -col=time,voltage -window=flattop -psdOutput
!sddsfft time.sdds hanning.sdds -col=time,voltage -window=hanning -psdOutput

# read sdds files
for window in ['none', 'flattop', 'hanning']:
    with open('%s.sdds' % window, 'r') as f:
        line = f.readline()
        while not line.startswith('&data'):
            line = f.readline()
        nRows = unpack('<l', f.read(4))[0]
        f.read(12)     # skip the two parameters (long, double)
        sddsfft = unpack('<%dd' % (nRows*3), f.read(8*3*nRows))
        sddsfft = [(sddsfft[i], sddsfft[i+1], sddsfft[i+1]**2, sddsfft[i+2]) for i in xrange(0, nRows*3, 3)]
    sdds[window] = np.array(sddsfft, dtype=[('f','f8'), ('fft','f8'), ('pwr','f8'), ('psd','f8')])
    
!ls -l *.sdds
```

### Plot comparison between Heinzel and SDDS for different windows

```python
for window in ['none', 'flattop', 'hanning']:
    plt.loglog(heinzel[window]['f'], heinzel[window]['fft'], alpha=0.5, label='Heinzel')
    plt.loglog(sdds[window]['f'], sdds[window]['fft']/np.sqrt(2), alpha=0.5, label='SDDS FFT/$\sqrt{2}$')
    plt.title('window: %s' % window)
    plt.ylabel('FFT (Vrms)')
    plt.xlabel('Frequency (Hz)')
    plt.legend()
    plt.show()
    for freq in [f1, f2]:
        print 'Heinzel %5d Hz %.3g Vrms' % (freq, heinzel[window]['fft'][heinzel[window]['f'] == freq][0])
        print 'SDDS    %5d Hz %.3g Vrms' % (freq, sdds[window]['fft'][sdds[window]['f'] == freq][0]/np.sqrt(2))
    print '-' * 60
```
